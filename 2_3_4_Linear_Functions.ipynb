```python
import matplotlib.pyplot as plt
import numpy as np

# Define the linear function
def linear_function(x):
    return 3 * x + 2

# Choose values for x
x_values = [0, 1]

# Calculate the corresponding y values for the chosen x values
y_values = [linear_function(x) for x in x_values]

# Plot the points
plt.figure(figsize=(10, 10))
plt.scatter(x_values, y_values, color='blue', s=100, label='Points: (0, 2), (1, 5)', zorder=5)  # Larger and coloured points

# Annotate the points
for i, (x, y) in enumerate(zip(x_values, y_values)):
    plt.annotate(f'({x}, {y})', (x, y), textcoords="offset points", xytext=(10, -10), fontsize=12)

# Plot the straight line through these points
x = np.linspace(-1, 2, 400)  # Extend the x-axis range for the line
y = linear_function(x)
plt.plot(x, y, color='green', linewidth=2, label='Linear Function: $f(x) = 3x + 2$')

# Set custom ticks and limits
plt.xticks(range(-1, 4), fontsize=12)
plt.yticks(range(0, 10), fontsize=12)
plt.xlim(-1, 2.5)
plt.ylim(0, 9)

# Add title and labels
plt.title('Linear Function', fontsize=16, weight='bold')
plt.xlabel('x', fontsize=14)
plt.ylabel('f(x)', fontsize=14)

# Add a legend
plt.legend(fontsize=12)

# Add a grid with lighter lines
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

# Highlight axes
plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', linewidth=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```
